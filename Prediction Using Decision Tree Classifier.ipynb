# Importing  Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.metrics as sm
%matplotlib inline

from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from matplotlib.pyplot import rcParams

import warnings
warnings.filterwarnings("ignore")
```

# READING DATA

```python
data = pd.read_csv("Iris.csv")
data.head()
```

```python
# checking null values
data.isnull().sum()
```

```python
data.describe()
```

```python
sns.pairplot(data, hue = 'Species')
```

# heatmap with the co-relation of the dataset

```python
sns.heatmap(data.corr(), annot= True)
```

# Splitting  and Processing Data

```python
X = data.iloc[:, [0,1,2,3]]
y = data.Species
data.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Decision Tree

```python
classifier = DecisionTreeClassifier(criterion='entropy')
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred_train = classifier.predict(X_train)
y_pred_test = classifier.predict(X_test)
```

```python
# let us print the accuracy score for the tree model
print("The accuracy score for the built model is:",accuracy_score(y_test, y_pred_test))
```

```python
rcParams['figure.figsize'] = 80, 20
plot_tree(classifier, precision=5, rounded = False, filled = True)
```
